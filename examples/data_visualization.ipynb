```python
import os
import pandas as pd
import matplotlib.pyplot as plt

def plot_by_query(file_list):
    all_results = []  # 用于保存所有文件的自动评判结果
    file_names = []  # 用于保存每个文件的文件名

    # 遍历文件列表
    for file_path in file_list:
        try:
            # 获取文件名
            file_name = os.path.basename(file_path)
            file_names.append(file_name)

            # 读取xlsx文件内容
            df = pd.read_excel(file_path)

            # 提取自动评判结果列（假设列名为'自动评判结果'，你需要替换为实际的列名）
            auto_judgment_results = df['自动评判结果']

            # 将自动评判结果添加到列表中
            all_results.append(auto_judgment_results.tolist())

        except Exception as e:
            print(f"Error reading file {file_path}: {e}")

    # 转换为DataFrame，行是文件名，列是每道题的结果
    df_results = pd.DataFrame(all_results, index=file_names)

    # 统计每个题的1和4的数量
    counts = df_results.apply(lambda x: x.value_counts()).fillna(0)

    # 打印结果
    print("统计结果:")
    print(counts)

    # 可以选择将统计结果可视化，这里使用了堆叠柱状图
    counts.T.plot(kind='bar', stacked=True)
    plt.title('Auto Judgment Results Count for Each Question')
    plt.xlabel('题目')
    plt.ylabel('数量')
    plt.legend(["值为1", "值为4"])
    plt.show()
    return counts

# 使用之前获取的文件列表进行读取和统计
result = plot_by_query(file_list)
```
