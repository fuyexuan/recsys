```python
import pandas as pd

# 定义 A_values 列表
A_values = ['不应拒答', '侵害他人合法权益', '公正歧视', '商业违法违规', '应拒答', '特定安全需求', '违反社会主义核心价值观']

# 创建一个空的数据框，用于存储合并后的数据
merged_df = pd.DataFrame()

# 循环处理每个 A_value 对应的文件
for A_value in A_values:
    print(A_value)
    # 指定 Excel 文件路径
    excel_file_path = f'{A_value}_ability.xlsx'

    # 使用 pandas 读取 Excel 文件
    df = pd.read_excel(excel_file_path)

    # 添加一列 A_value
    df['area'] = A_value

    # 将当前数据框与已合并的数据框进行合并
    merged_df = pd.concat([merged_df, df], ignore_index=True)

# 打印合并后的数据框
print(merged_df)

```

```python
import os
def get_all_files_in_folder(folder_path, prefix=''):
    all_files = []

    # 遍历文件夹中的所有文件和子文件夹
    for root, dirs, files in os.walk(folder_path):
        for file_name in files:
            # 构建文件的完整路径
            file_path = os.path.join(root, file_name)
            # 检查文件名是否以指定前缀开头
            if file_name.startswith(prefix):
                # 将文件路径添加到列表中
                all_files.append(file_path)

    return all_files

folder_path = r'C:\Users\17235\Desktop\fyx\LLM-Trade\recsys_linux\0102测试结果'  # 替换为你的文件夹路径

# 获取文件夹及其所有子文件夹中以'AAA'开头的所有文件
file_list = get_all_files_in_folder(folder_path, prefix=A_values[0])

# 打印文件列表
for file_path in file_list:
    print(file_path)

print(subject_names)
```
